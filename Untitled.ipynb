```python
columns_to_drop = ['iso_code', 'GDLCODE','level','sgdi','shdif','shdim','healthindex','healthindexf','healthindexm','incindex', 'incindexf', 'incindexm','edindexf', 'edindexm','eschf', 'eschm','mschf', 'mschm','lifexpf', 'lifexpm','gnicf','gnicm','lgnic', 'lgnicf', 'lgnicm', 'pop']
shdi.drop(columns=columns_to_drop, inplace=True)
```

```python
shdi['shdi'] = pd.to_numeric(shdi['shdi'], errors='coerce')
shdi['edindex'] = pd.to_numeric(shdi['edindex'], errors='coerce')
shdi['esch'] = pd.to_numeric(shdi['esch'], errors='coerce')
shdi['msch'] = pd.to_numeric(shdi['msch'], errors='coerce')
shdi['lifexp'] = pd.to_numeric(shdi['lifexp'], errors='coerce')
shdi['gnic'] = pd.to_numeric(shdi['gnic'], errors='coerce')
```

```python
def categorise_hdi(row):
    shdi_value = row['shdi']
    if (shdi_value >= 0.8):
        return "Very High"
    elif(shdi_value>=0.7):
        return "High"
    elif(shdi_value>=0.550):
        return "Medium"
    else:
        return "Low"
shdi['hdi_category'] = shdi.apply(categorise_hdi, axis=1)
```

```python
df = shdi[(shdi["country"]=='India') & (shdi['region']=='Total')]
df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score
from statsmodels.tsa.arima.model import ARIMA

features = df[['year', 'edindex','esch','msch','lifexp','gnic']]


train_size = int(len(df) * 0.8)
train, test = df[df["year"]!=2021], df[df["year"]==2021]

arima_model = ARIMA(train['shdi'], order=(5,1,0))
arima_fit = arima_model.fit()

arima_predictions = arima_fit.forecast(steps=len(test)+5)




X = features
y = df['shdi']

X_train, X_test, y_train, y_test = features[features["year"]!=2021] , features[features["year"]==2021] , df[df['year']!=2021]['shdi'],df[df['year']==2021]['shdi']

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
print(f'Random Forest RMSE: {rf_rmse}')


categories = df['hdi_category']

X_class_train, X_class_test, y_class_train, y_class_test = train_test_split(X, categories, test_size=0.2, random_state=42)

rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_class_train, y_class_train)

class_predictions = rf_classifier.predict(X_class_test)

class_accuracy = accuracy_score(y_class_test, class_predictions)
print(f'Classification Accuracy: {class_accuracy}')
```
